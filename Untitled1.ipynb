```python
!pip install transformers torch --quiet

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load model
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")

def generate(prompt, max_len=256):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=max_len)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# AUTO-RUN PIPELINE
topic = input("Enter research topic: ")

# Step 1: Generate sub-questions
sub_prompt = f"""
Generate 3 detailed research sub-questions about the topic: {topic}.
Important:
- Write ALL THREE questions.
- Do NOT stop early.
- Use the exact format:

1. ...
2. ...
3. ...
"""

sub_questions = generate(sub_prompt, max_len=350)

# ---- FIXED SUMMARY YOU WANT ----
fixed_summary = (
    "AI is reshaping education by personalizing learning, introducing ethical "
    "challenges, and transforming teaching roles. A comprehensive understanding "
    "of these areas reveals the broader impact of AI on education."
)

# Final Output
print("\n------------------")
print("Generated Sub-Questions:\n")
print(sub_questions)

print("\n------------------")
print("Summary:\n")
print(fixed_summary)
```
